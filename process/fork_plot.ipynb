```python
import networkx as nx
import matplotlib.pyplot as plt

def visualize_tree(text:str):

    lines = text.strip().split('\n')
    output = [(line.split(' ----> ')[0], line.split(' ----> ')[1]) for line in lines]

    G = nx.DiGraph()

    for parent, child in output:
        parent_id = int(parent.split(": ")[-1])
        child_id = int(child.split(": ")[-1])
        G.add_edge(parent_id, child_id)

    pos = nx.spring_layout(G)
    labels = {node: f"PID:{node}" for node in G.nodes}

    nx.draw(G, pos, with_labels=True, labels=labels, font_size=6, node_size=1000, node_color="skyblue", font_color="black", font_weight="bold", edge_color="gray", arrowsize=20)
    plt.show()
```

```python
import subprocess
```

```python
deep_2 = subprocess.check_output(['python', 'fork_finite.py', '2'], text=True)
deep_2
```

```python
visualize_tree(deep_2)
```

```python
deep_3 = subprocess.check_output(['python', 'fork_finite.py', '3'], text=True)
deep_3
```

```python
visualize_tree(deep_3)
```

```python
deep_4 = subprocess.check_output(['python', 'fork_finite.py', '4'], text=True)
deep_4
```

```python
visualize_tree(deep_4)
```
